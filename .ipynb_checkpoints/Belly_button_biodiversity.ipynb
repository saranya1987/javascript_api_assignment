```python
import pandas as pd
import numpy as np
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, desc
from flask import Flask, jsonify, render_template, request
import csv
```

```python
engine = create_engine("sqlite:///db/belly_button_biodiversity.sqlite")
Base = automap_base()

Base.prepare(engine, reflect=True)

Otu = Base.classes.otu
Samples = Base.classes.samples
Samples_Metadata = Base.classes.samples_metadata

session = Session(engine)
```

```python
stmt = session.query(Otu.lowest_taxonomic_unit_found).all()
otu_list = list(np.ravel(stmt))

print(otu_list)
```

```python
sample_ids = []

results = session.query(Samples_metadata.SAMPLEID)

for result in results:
    sample_ids.append("BB_" + str(result[0]))
sample_ids
```

```python
Otu_desc = []

results = session.query(Otu.lowest_taxonomic_unit_found)

for result in results:
    Otu_desc.append(str(result[0]))
Otu_desc
```

```python
sel = [Samples_Metadata.SAMPLEID, Samples_Metadata.ETHNICITY,
           Samples_Metadata.GENDER, Samples_Metadata.AGE,
           Samples_Metadata.LOCATION, Samples_Metadata.BBTYPE]
results = session.query(*sel).\
        filter(Samples_Metadata.SAMPLEID).all()
print(results)
```
